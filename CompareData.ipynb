# Obtención de las tablas que conforman el dataframe 

```python
def tablas_en_csv(csv,*, verbose=False):
    '''
    Funcion para visualiar el numero de tanlas en el dataframe 
    
    Retorna los nombres de las tablas
    '''
    llaves = list(csv.keys())
    if type(llaves[0]) == tuple:
        llaves = np.array(llaves)
        tablas = set(llaves[:,0])
        if verbose:
            print(f'Hay {len(tablas)} tablas')
        return list(tablas)
    else:
        print(f'El dataset contiene una unica tabla')
```

# Unir las diferentes caracterisiticas 

Lo voy a unir diferenciando entre categoricas y numericas 

```python
cateList = dict() #Diccionario para almacenar las categorizaciones
```

```python
def categorizacion(col, colname):
    valores = col.value_counts()
    cate = np.zeros(len(col), dtype=int)
    llaves = list(valores.keys())
    l = []
    if type(col[0]) == bool:
        print("es bool")
        cate = list(map(int,col))
    else:
        for i, key in enumerate(llaves):
            cate[col == key] = i
            l.append((key,i))
    if colname in cateList:
        print("La columna ya ha sido categorizada con anterioridad")
    else:
        cateList[colname] = l
    return pd.Series(cate)
```
